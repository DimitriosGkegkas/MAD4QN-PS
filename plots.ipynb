```python
travel_time_histo = {
    'FTTL2': [np.mean(travel_time_fttl2), np.mean(waiting_time_fttl2)],
    'FTTL1': [np.mean(travel_time_fttl1), np.mean(waiting_time_fttl1)],
    'ATL2': [np.mean(travel_time_atl2), np.mean(waiting_time_atl2)],
    'ATL1': [np.mean(travel_time_atl1), np.mean(waiting_time_atl1)],
    'RP': [np.mean(travel_time_random), np.mean(waiting_time_random)],
    'MAD4QN-PS': [np.mean(travel_time_mad4qn), np.mean(waiting_time_mad4qn)]
}

travel_time_histo2 = {
    'FTTL1': [np.mean(travel_time_fttl1), np.mean(waiting_time_fttl1)],
    'FTTL2': [np.mean(travel_time_fttl2), np.mean(waiting_time_fttl2)],
    'ATL1': [np.mean(travel_time_atl1), np.mean(waiting_time_atl1)],
    'ATL2': [np.mean(travel_time_atl2), np.mean(waiting_time_atl2)],
    'RP': [np.mean(travel_time_random), np.mean(waiting_time_random)],
    'MAD4QN-PS': [np.mean(travel_time_mad4qn), np.mean(waiting_time_mad4qn)]
}

x = range(0, 6)
bar_width = 0.4

facecolors = ['green', 'darkblue', 'gray', 'red']
edgecolors = facecolors

for i, pair in enumerate(travel_time_histo2.values()):
    if i in (0, 1):
        plt.bar(x[i] - bar_width/2, pair[0], bar_width, color=facecolors[1], edgecolor=edgecolors[1], alpha=0.8)
        plt.bar(x[i] + bar_width/2, pair[1], bar_width, color=facecolors[1], edgecolor=edgecolors[1], alpha=0.3)        
    elif i in (2, 3):
        plt.bar(x[i] - bar_width/2, pair[0], bar_width, color=facecolors[1], edgecolor=edgecolors[1], alpha=0.8,
               )
        plt.bar(x[i] + bar_width/2, pair[1], bar_width, color=facecolors[1], edgecolor=edgecolors[1], alpha=0.3)        
    elif i == 4:
        plt.bar(x[i] - bar_width/2, pair[0], bar_width, color=facecolors[1], edgecolor=edgecolors[1], alpha=0.8,
               label='Average travel time per vehicle')
        plt.bar(x[i] + bar_width/2, pair[1], bar_width, color=facecolors[1], edgecolor=edgecolors[1], alpha=0.3,
               label='Average waiting time per vehicle')        
    else:
        plt.bar(x[i] - bar_width/2, pair[0], bar_width, color=facecolors[1], edgecolor=edgecolors[1], alpha=0.8)
        plt.bar(x[i] + bar_width/2, pair[1], bar_width, color=facecolors[1], edgecolor=edgecolors[1], alpha=0.3)        
plt.xticks(range(0, 6), list(travel_time_histo2.keys()), fontsize=14)
# plt.xlabel('Method', fontsize=14)
plt.ylabel('Time [s]', fontsize=14)
plt.yticks(fontsize=14)
plt.legend(loc='upper right', fontsize=12)
plt.grid(axis='y')
plt.savefig("travel_time.pdf", format='pdf')  
plt.show()
```

```python
collisions_histo = {
    'RP': np.mean(collisions_random),
    'MAD4QN-PS': np.mean(collisions_mad4qn)
}

x = [0, 0.25]
bar_width = 0.15

facecolors = ['green', 'darkgreen', 'gray', 'red']
edgecolors = facecolors

for i, pair in enumerate(collisions_histo.values()):
    if i == 0:
        plt.bar(x[i], pair*100, bar_width, color=facecolors[1], edgecolor=edgecolors[1], alpha=0.8)
    else:
        plt.bar(x[i], pair*100, bar_width, color=facecolors[1], edgecolor=edgecolors[1], alpha=0.8)
plt.xticks([0, 0.25], list(collisions_histo.keys()), fontsize=14)
plt.ylabel('Percentage [%]', fontsize=14)
plt.yticks(fontsize=14)
plt.grid(axis='y')
plt.savefig("collisions.png", format='png')  
plt.show()
```
